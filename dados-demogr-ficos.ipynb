```python
contagem_generos = adultos['sex'].value_counts()

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.pie(contagem_generos, labels=contagem_generos.index, autopct='%1.1f%%')
plt.title('Distribuição por Gênero')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
contagem_profissoes = Profissao.value_counts()

# Criando o gráfico de barras
plt.figure(figsize=(10,6))
sns.barplot(x=contagem_profissoes.index, y=contagem_profissoes)
plt.title('Distribuição das Profissões')
plt.xlabel('Profissão')
plt.ylabel('Frequência')
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x para melhor visualização
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.color_palette("husl")
sns.set_style("darkgrid")

sns.countplot(x='education', data=adultos)
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Contagem')
plt.title('Distribuição dos Níveis de Escolaridade por Gênero', fontsize=16)
plt.xticks(rotation=45)
plt.show()
```
