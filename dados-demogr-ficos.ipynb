```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
contagem_profissoes = Profissao.value_counts()

# Criando o grfico de barras
plt.figure(figsize=(10,6))
sns.barplot(x=contagem_profissoes.index, y=contagem_profissoes)
plt.title('Distribuio das Profisses')
plt.xlabel('Profisso')
plt.ylabel('Frequncia')
plt.xticks(rotation=45)  # Rotacionar os rtulos do eixo x para melhor visualizao
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.color_palette("husl")
sns.set_style("darkgrid")

sns.countplot(x='education', data=adultos)
plt.xlabel('Nvel de Escolaridade')
plt.ylabel('Contagem')
plt.title('Distribuio dos Nveis de Escolaridade por Gnero', fontsize=16)
plt.xticks(rotation=45)
plt.show()
```
